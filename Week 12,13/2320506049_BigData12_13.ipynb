```python
# Exercise 1: Identifying and Handling Missing Data
import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age': [24, 30, None, 22, 35],
    'Salary': [48000, None, 57000, None, 60000]
}
df = pd.DataFrame(data)

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
df = df.dropna(subset=['Name'])
print('After cleaning:\n', df)

```

```python
# Exercise 2: Standardizing Categorical Data
# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}
df = pd.DataFrame(data)

# Standardize category values
df['Category'] = df['Category'].str.capitalize()
print('Standardized Data:\n', df)
```

```python
import matplotlib.pyplot as plt
import numpy as np  # Pastikan numpy diimpor

# Data for plotting
categories = ['Category A', 'Category B', 'Category C']
group1_values = [23, 17, 35]
group2_values = [30, 25, 40]

# Bar width
bar_width = 0.35

# Set positions for bars
index = np.arange(len(categories))  # Menggunakan numpy untuk membuat array index

# Plot
plt.figure(figsize=(10, 6))
plt.bar(index, group1_values, bar_width, label='Group 1', color='cyan')
plt.bar(index + bar_width, group2_values, bar_width, label='Group 2', color='orange')

# Customize plot
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Multi-Bar Chart for Comparing Groups')
plt.xticks(index + bar_width / 2, categories)  # Set category labels to the center
plt.legend()

# Show plot
plt.show()
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv('taxi_trip_pricing.csv')

# Menampilkan informasi awal
print(df.info())

# Identifikasi missing values
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalisasi kolom numerik
scaler = MinMaxScaler()
df[['Trip_Distance_km', 'Base_Fare', 'Per_Km_Rate', 'Per_Minute_Rate', 'Trip_Duration_Minutes', 'Trip_Price']] = scaler.fit_transform(
    df[['Trip_Distance_km', 'Base_Fare', 'Per_Km_Rate', 'Per_Minute_Rate', 'Trip_Duration_Minutes', 'Trip_Price']])

print("\nData setelah normalisasi:")
print(df.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Standarisasi kolom kategorikal
encoder = LabelEncoder()
df['Time_of_Day'] = encoder.fit_transform(df['Time_of_Day'])
df['Day_of_Week'] = encoder.fit_transform(df['Day_of_Week'])
df['Traffic_Conditions'] = encoder.fit_transform(df['Traffic_Conditions'])
df['Weather'] = encoder.fit_transform(df['Weather'])

# Menghapus duplikasi
df.drop_duplicates(inplace=True)

print("\nData setelah standarisasi dan duplikasi dihapus:")
print(df.head())
```

```python
df.interpolate(method='linear', inplace=True)
print("\nJumlah nilai yang hilang setelah interpolasi:")
print(df.isnull().sum())
```

```python
# Fungsi untuk menangani outliers dengan IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df = remove_outliers(df, 'Trip_Price')
df = remove_outliers(df, 'Trip_Distance_km')
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribusi harga trip
plt.figure(figsize=(10, 6))
sns.histplot(df['Trip_Price'], kde=True, color='blue')
plt.title('Distribusi Harga Trip')
plt.xlabel('Harga Trip')
plt.ylabel('Frekuensi')
plt.show()

# Distribusi jarak trip
plt.figure(figsize=(10, 6))
sns.histplot(df['Trip_Distance_km'], kde=True, color='green')
plt.title('Distribusi Jarak Trip')
plt.xlabel('Jarak Trip (km)')
plt.ylabel('Frekuensi')
plt.show()
```
